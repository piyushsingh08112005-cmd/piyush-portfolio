## Source Code

```
#include <stdio.h>

double add(double a, double b) { return a + b; }
double sub(double a, double b) { return a - b; }
double mul(double a, double b) { return a * b; }
double divide(double a, double b) { return b != 0 ? a / b : 0; }

int main(void) {
    int choice;
    double a, b;
    do {
        printf("\n===== Calculator =====\n");
        printf("1. Add\n2. Subtract\n3. Multiply\n4. Divide\n0. Exit\nChoice: ");
        if (scanf("%d", &choice) != 1) return 0;
        if (choice == 0) break;
        printf("Enter two numbers: ");
        scanf("%lf %lf", &a, &b);
        switch (choice) {
            case 1: printf("Result = %.2f\n", add(a, b)); break;
            case 2: printf("Result = %.2f\n", sub(a, b)); break;
            case 3: printf("Result = %.2f\n", mul(a, b)); break;
            case 4: printf("Result = %.2f\n", divide(a, b)); break;
            default: printf("Invalid choice\n");
        }
    } while (choice != 0);
    return 0;
}
```
