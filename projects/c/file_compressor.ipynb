## Source Code

```
#include <stdio.h>
#include <string.h>

void compress(const char *input, const char *output) {
    FILE *in = fopen(input, "r");
    FILE *out = fopen(output, "w");
    int ch, prev = EOF, count = 0;
    if (!in || !out) return;
    while ((ch = fgetc(in)) != EOF) {
        if (ch == prev) {
            count++;
        } else {
            if (prev != EOF) fprintf(out, "%d%c", count, prev);
            prev = ch;
            count = 1;
        }
    }
    if (prev != EOF) fprintf(out, "%d%c", count, prev);
    fclose(in);
    fclose(out);
}

void decompress(const char *input, const char *output) {
    FILE *in = fopen(input, "r");
    FILE *out = fopen(output, "w");
    int count;
    char ch;
    if (!in || !out) return;
    while (fscanf(in, "%d%c", &count, &ch) == 2) {
        for (int i = 0; i < count; i++) fputc(ch, out);
    }
    fclose(in);
    fclose(out);
}

int main(void) {
    compress("input.txt", "compressed.txt");
    decompress("compressed.txt", "output.txt");
    printf("Compression and decompression completed.\n");
    return 0;
}
```
