## Source Code

```
#include <iostream>
#include <iomanip>
using namespace std;

class Account {
private:
    string holder;
    int pin;
    double balance;

public:
    Account(string h, int p, double b) : holder(move(h)), pin(p), balance(b) {}
    bool login(int entered) const { return entered == pin; }
    void deposit(double amt) { balance += amt; }
    bool withdraw(double amt) { if (amt > balance) return false; balance -= amt; return true; }
    void show() const { cout << holder << " | Balance: " << fixed << setprecision(2) << balance << endl; }
};

int main() {
    Account acc("Piyush Singh", 1234, 15000.00);
    int pin;
    cout << "Enter PIN: ";
    cin >> pin;
    if (!acc.login(pin)) { cout << "Wrong PIN\n"; return 0; }

    int choice;
    do {
        cout << "\n1.Deposit 2.Withdraw 3.Balance 0.Exit\nChoice: ";
        cin >> choice;
        if (choice == 1) {
            double amt; cin >> amt; acc.deposit(amt);
        } else if (choice == 2) {
            double amt; cin >> amt;
            if (!acc.withdraw(amt)) cout << "Insufficient funds\n";
        } else if (choice == 3) {
            acc.show();
        }
    } while (choice != 0);
    return 0;
}
```
