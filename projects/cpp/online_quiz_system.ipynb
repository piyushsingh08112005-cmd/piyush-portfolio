## Source Code

```
#include <iostream>
#include <vector>
#include <cctype>
#include <string>
#include <utility>
using namespace std;

struct Question {
    string text, optionA, optionB, optionC, optionD;
    char answer;
};

int main() {
    vector<Question> quiz = {
        {"What is the output of 2 + 2?", "3", "4", "5", "6", 'B'},
        {"Which keyword creates a class in C++?", "struct", "class", "object", "def", 'B'}
    };

    int score = 0;
    for (size_t i = 0; i < quiz.size(); ++i) {
        cout << "\nQ" << i + 1 << ". " << quiz[i].text << '\n';
        cout << "A. " << quiz[i].optionA << '\n';
        cout << "B. " << quiz[i].optionB << '\n';
        cout << "C. " << quiz[i].optionC << '\n';
        cout << "D. " << quiz[i].optionD << '\n';
        char ans;
        cin >> ans;
        if (toupper(ans) == quiz[i].answer) score++;
    }
    cout << "\nFinal Score: " << score << "/" << quiz.size() << endl;
    return 0;
}
```
