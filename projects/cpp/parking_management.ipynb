## Source Code

```
#include <iostream>
#include <vector>
#include <string>
#include <iomanip>
#include <algorithm>
#include <utility>

using namespace std;

class Vehicle {
private:
    int id;
    string name;
    string meta1;
    string meta2;
    double amount;
    int quantity;

public:
    Vehicle() : id(0), amount(0.0), quantity(0) {}
    Vehicle(int id, string name, string meta1, string meta2, double amount, int quantity)
        : id(id), name(move(name)), meta1(move(meta1)), meta2(move(meta2)), amount(amount), quantity(quantity) {}

    int getId() const { return id; }
    string getName() const { return name; }

    void display() const {
        cout << left << setw(5) << id
             << setw(22) << name.substr(0, 21)
             << setw(18) << meta1.substr(0, 17)
             << setw(18) << meta2.substr(0, 17)
             << setw(12) << fixed << setprecision(2) << amount
             << setw(8) << quantity << '\n';
    }

    void updateAmount(double v) { amount = v; }
    void updateQuantity(int v) { quantity = v; }
    void updateText(int slot, const string& v) {
        if (slot == 1) meta1 = v;
        else if (slot == 2) meta2 = v;
        else name = v;
    }
};

class ParkingManager {
private:
    vector<Vehicle> items;
    int nextId = 1;

public:
    void add(const string& name, const string& meta1, const string& meta2, double amount, int quantity) {
        items.emplace_back(nextId++, name, meta1, meta2, amount, quantity);
    }

    void list() const {
        cout << "\nID   NAME                  INFO 1            INFO 2            AMOUNT      QTY\n";
        cout << "-------------------------------------------------------------------------------\n";
        for (const auto& item : items) item.display();
    }

    Vehicle* find(int id) {
        for (auto& item : items) if (item.getId() == id) return &item;
        return nullptr;
    }

    void remove(int id) {
        items.erase(remove_if(items.begin(), items.end(),
                              [id](const Vehicle& item) { return item.getId() == id; }),
                    items.end());
    }
};

int main() {
    ParkingManager manager;
    manager.add("Parking Management Alpha", "Priority", "Active", 1200.00, 10);
    manager.add("Parking Management Beta", "Standard", "Available", 950.50, 5);

    int choice;
    do {
        cout << "\n===== Parking Management =====\n";
        cout << "1. Add\n2. List\n3. Update\n4. Delete\n0. Exit\nChoice: ";
        cin >> choice;

        if (choice == 1) {
            string name, a, b; double amount; int qty;
            cin.ignore();
            cout << "Name: "; getline(cin, name);
            cout << "Info 1: "; getline(cin, a);
            cout << "Info 2: "; getline(cin, b);
            cout << "Amount: "; cin >> amount;
            cout << "Quantity: "; cin >> qty;
            manager.add(name, a, b, amount, qty);
        } else if (choice == 2) {
            manager.list();
        } else if (choice == 3) {
            int id, field;
            cout << "ID: "; cin >> id;
            auto* item = manager.find(id);
            if (!item) {
                cout << "Not found.\n";
                continue;
            }
            cout << "1.Name 2.Info1 3.Info2 4.Amount 5.Quantity: ";
            cin >> field;
            cin.ignore();
            if (field <= 3) {
                string value;
                getline(cin, value);
                item->updateText(field, value);
            } else if (field == 4) {
                double v; cin >> v;
                item->updateAmount(v);
            } else if (field == 5) {
                int v; cin >> v;
                item->updateQuantity(v);
            }
        } else if (choice == 4) {
            int id; cout << "ID: "; cin >> id; manager.remove(id);
        }
    } while (choice != 0);

    return 0;
}
```
