## Source Code

```
#include <iostream>
#include <vector>
#include <cctype>
#include <cstdlib>
#include <ctime>
using namespace std;

struct Point { int x, y; };

bool collide(const vector<Point>& snake, int x, int y) {
    for (const auto& s : snake) if (s.x == x && s.y == y) return true;
    return false;
}

int main() {
    const int W = 20, H = 10;
    vector<Point> snake = {{10, 5}, {9, 5}, {8, 5}};
    Point food{3, 3};
    char dir = 'R';
    int score = 0;
    srand((unsigned)time(nullptr));

    cout << "Terminal Snake (simple turn-based version)\n";
    cout << "Use W A S D then Enter each move.\n";

    for (;;) {
        cout << "\nScore: " << score << "\n";
        for (int y = 0; y < H; ++y) {
            for (int x = 0; x < W; ++x) {
                bool drawn = false;
                if (food.x == x && food.y == y) { cout << '*'; continue; }
                for (size_t i = 0; i < snake.size(); ++i) {
                    if (snake[i].x == x && snake[i].y == y) { cout << (i == 0 ? 'O' : 'o'); drawn = true; break; }
                }
                if (!drawn) cout << '.';
            }
            cout << '\n';
        }

        char ch;
        cout << "Move: ";
        cin >> ch;
        ch = toupper(ch);
        if (ch == 'W' || ch == 'A' || ch == 'S' || ch == 'D') dir = ch;

        Point head = snake.front();
        if (dir == 'W') head.y--;
        if (dir == 'S') head.y++;
        if (dir == 'A') head.x--;
        if (dir == 'D') head.x++;

        if (head.x < 0 || head.x >= W || head.y < 0 || head.y >= H || collide(snake, head.x, head.y)) {
            cout << "Game Over! Final score: " << score << endl;
            break;
        }

        snake.insert(snake.begin(), head);
        if (head.x == food.x && head.y == food.y) {
            score++;
            food = {rand() % W, rand() % H};
        } else {
            snake.pop_back();
        }
    }
    return 0;
}
```
